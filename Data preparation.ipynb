# Import Libraries

# Data3

```python
data3 = pd.read_csv('./datasets/labeled_data.csv')
data3
```

```python
data3.drop(['count','hate_speech','offensive_language','neither','Unnamed: 0'], inplace=True, axis=1)
data3
```

```python
data3.rename(columns = {'tweet' : 'Text', 'class' : 'Label'}, inplace=True)
data3
```

```python
data3.dropna(inplace=True)
data3.shape
```

```python
data3.drop_duplicates(inplace=True)
data3.shape
```

```python
data3 = data3.reindex(columns=['Text', 'Label'])
data3
```

```python
data3['Label'].replace({0:2, 1:3, 2:0}, inplace=True)
```

```python
data3
```

# Data4

```python
data4 = pd.read_csv('./datasets/DynamicallyGeneratedHateDataset-entries-v0.1.csv')
data4
```

```python
data4 = data4[['text', 'label']]
data4
```

```python
data4 = data4[data4['label'] == 'hate']
data4
```

```python
data4['label'].replace({'hate' : 2}, inplace=True)
data4
```

```python
data4.rename(columns={'text' : 'Text', 'label':'Label'}, inplace=True)
data4
```

```python
_, splitdata4 = train_test_split(data4, test_size=0.5)
```

```python
splitdata4.dropna(inplace=True)
splitdata4.drop_duplicates(inplace=True)
splitdata4.shape
```

# Data6

```python
data6 = pd.read_csv('./datasets/Xenophobia.csv', encoding='ANSI')
data6
```

```python
data6 = data6[['text']]
data6
```

```python
data6.dropna(inplace=True)
data6.shape
```

# Data7

```python
data7 = pd.read_csv('./datasets/train.csv')
data7
```

```python
data7 = data7[['tweet', 'label']]
data7
```

```python
data7 = data7[data7.label == 1]
data7
```

# Final Data

```python
frames = [data1, data2, data3, splitdata4, data5, splitdata6, data7]
data = pd.concat(frames)
data
```
